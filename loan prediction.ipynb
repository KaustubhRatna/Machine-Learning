```python
X_train, X_test, y_train, y_test = train_test_split(df1, Y_train, test_size=0.3, random_state=0)

X_train["Gender"]=X_train["Gender"].replace('Male',1)
X_train["Gender"]=X_train["Gender"].replace('Female',0)

X_train["Married"]=X_train["Married"].replace('Yes',1)
X_train["Married"]=X_train["Married"].replace('No',0)

X_train["Dependents"]=X_train["Dependents"].replace('1',1)
X_train["Dependents"]=X_train["Dependents"].replace('0',0)
X_train["Dependents"]=X_train["Dependents"].replace('2',2)
X_train["Dependents"]=X_train["Dependents"].replace('3+',3)

X_train["Education"]=X_train["Education"].replace('Graduate',1)
X_train["Education"]=X_train["Education"].replace('Not Graduate',0)

X_train["Self_Employed"]=X_train["Self_Employed"].replace('No',0)
X_train["Self_Employed"]=X_train["Self_Employed"].replace('Yes',1)

X_train["Property_Area"]=X_train["Property_Area"].replace('Urban',3)
X_train["Property_Area"]=X_train["Property_Area"].replace('Rural',1)
X_train["Property_Area"]=X_train["Property_Area"].replace('Semiurban',2)


```

```python
X_train=X_train.drop(['Loan_ID'],axis=1)
```

```python
X_train.head()
```

```python
cols=list(X_train.columns)
cols.remove('ApplicantIncome')
```

```python
cols
```

```python


X_train.head()

#y_train.head()
```

```python
logreg = LogisticRegression()
logreg.fit(X_train[cols], y_train)
```

```python
## importance plot
plt.figure(figsize=(20,8))
pd.Series(np.std(X_train[cols])*logreg.coef_.flatten()).sort_values(ascending=False).plot.bar()
plt.show()
```

```python
y_pred = logreg.predict(X_train[cols])
print('Accuracy of logistic regression classifier on train set: {:.2f}'.format(logreg.score(X_train[cols], y_train)))
```
